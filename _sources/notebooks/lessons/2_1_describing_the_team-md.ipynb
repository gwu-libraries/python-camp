## Tabular data

A single dictionary is a good way to represent something in the real world -- like a person, a course, or a textbook -- that has multiple properties or attributes. 

The dictionary's keys can help us organize the attributes by labels, just the way we might label the different elements of each entry in an addressbook: name, email address, phone number, etc.

You might also think of the dictionary keys as the column headings in a spreadsheet. In a spreadsheet, each row uses the same column headings, which allows us to do things like sort or filter the data by particular attributes. For example, if we covert our addressbook to an Excel file, assuming we have a column for email address, we can easily compute how many of our contacts have GW email addresses.

We call such a data structure **tabular** because it is organized as a table, i.e., with rows and columns.

### Tabular data in Python

How do we represent tabular data in Python? 

There are some very handy Python libraries for working with tables, [pandas](https://pandas.pydata.org/) being one of the most popular. But we can also handle tabular data with just the Python types you've already encountered. 

One common approach, which we will work through together, is to create a **list of dictionaries**. 

- On the analogy with a spreadsheet, each dictionary in the list would be another row of the table. 
- Each dictionary would have the same **keys** (with different **values**), corresponding to the data in the various columns of the spreadsheet.
